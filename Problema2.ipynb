```python
import pandas as pd

df = pd.read_csv('./data/winemag-data-130k-v2.csv')

print(df.info())
print(df.head())

df = df.rename(columns={'columna_original_1': 'nueva_columna_1',
                        'columna_original_2': 'nueva_columna_2',
                        'columna_original_3': 'nueva_columna_3',
                        'columna_original_4': 'nueva_columna_4'})

continentes = {'Spain': 'Europa', 'France': 'Europa', 'Italy': 'Europa', 'Argentina': 'América', 'Chile': 'América',
               'US': 'América', 'Australia': 'Oceanía', 'New Zealand': 'Oceanía'}
df['continente'] = df['country'].map(continentes)

reporte_1 = df.groupby('continente')['points'].mean()
reporte_2 = df.groupby('country')['price'].max()
reporte_3 = df.groupby(['country', 'variety']).size().reset_index(name='cantidad_vinos')
reporte_4 = df.groupby('country').agg({'price': 'mean', 'points': 'count'}).sort_values(by='price', ascending=False)

reporte_4.to_csv('reporte4.csv', index=False)

print(reporte_1)
print(reporte_2)
print(reporte_3)
print(reporte_4)
```
